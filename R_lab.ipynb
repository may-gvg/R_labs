<- (operator przypisania):
Jest to tradycyjny operator przypisania w R.
Może być używany zarówno w konsoli, jak i wewnątrz funkcji.
Jest bardziej uniwersalny i często preferowany przez społeczność R dla czytelności, zwłaszcza w skryptach i funkcjach.
Przykład:
x <- 5

= (operator równości):
Choć może być używany do przypisywania wartości w globalnym środowisku (np. w konsoli), jego głównym zastosowaniem jest przypisywanie wartości do argumentów funkcji podczas jej wywoływania.
Zaleca się ostrożność przy używaniu = w innych kontekstach niż argumenty funkcji, aby unikać niejednoznaczności.
rzykład użycia w kontekście argumentów funkcji:

plot(x = 1:10, y = 1:10)

Podsumowując, choć oba znaki mogą być używane do przypisywania wartości, 
<- jest częściej stosowany do ogólnego przypisywania, podczas gdy = jest głównie używany do przypisywania wartości argumentów funkcji. 
W praktyce zaleca się stosowanie <- do przypisywania wartości poza kontekstem argumentów funkcji, aby zapewnić czytelność i spójność kodu.

```r
# Zakładanie wektorów

# 1. Tworzenie wektora numerycznego
num_vector <- c(1, 2, 3, 4, 5)
print("Numerical vector:")
print(num_vector)

# 2. Tworzenie wektora znakowego
char_vector <- c("apple", "banana", "cherry")
print("Character vector:")
print(char_vector)

# 3. Tworzenie wektora logicznego
logic_vector <- c(TRUE, FALSE, TRUE, FALSE)
print("Logical vector:")
print(logic_vector)

# 4. Nadawanie nazw elementom wektora
names(num_vector) <- c("first", "second", "third", "fourth", "fifth")
print("Named numerical vector:")
print(num_vector)

# Operacje na wektorach

# 1. Długość wektora
print("Length of the numerical vector:")
print(length(num_vector))

# 2. Suma elementów wektora
print("Sum of elements in the numerical vector:")
print(sum(num_vector))

# 3. Średnia wartość elementów wektora
print("Average value of the numerical vector:")
print(mean(num_vector))

# 4. Mediana wartości wektora
print("Median value of the numerical vector:")
print(median(num_vector))

# 5. Sortowanie wektora
sorted_vector <- sort(num_vector)
print("Sorted numerical vector:")
print(sorted_vector)

# 6. Wybieranie elementów wektora
print("Second element of the numerical vector:")
print(num_vector[2])

# 7. Tworzenie sekwencji
seq_vector <- seq(1, 10, by=2)
print("Sequence vector:")
print(seq_vector)

# 8. Powtarzanie elementów
repeat_vector <- rep(5, times=3)
print("Repeated vector:")
print(repeat_vector)

# 9. Operacje z funkcją order
vec <- c(100, 15, 300, 40)
ordered_indices <- order(vec)
print("Indices that will sort the vector:")
print(ordered_indices)
print("Sorted vector using the ordered indices:")
print(vec[ordered_indices])

# 10. Operacje z funkcją rank
vec2 <- c(100, 15, 100, 40)
ranking <- rank(vec2)
print("Ranking of the elements in the vector:")
print(ranking)

# Operacje na wektorach logicznych

# 1. Porównania
comparison <- num_vector > 3
print("Comparison result:")
print(comparison)

# 2. Wszystkie elementy prawdziwe?
print("Are all elements TRUE?")
print(all(comparison))

# 3. Czy jakikolwiek element jest prawdziwy?
print("Is any element TRUE?")
print(any(comparison))

# 4. Wybieranie elementów na podstawie wektora logicznego
print("Elements greater than 3:")
print(num_vector[comparison])

# Algebraiczne operacje na wektorach

# 1. Dodawanie wektorów
print("Adding vectors:")
print(num_vector + c(1, 1, 1, 1, 1))

# 2. Odejmowanie wektorów
print("Subtracting vectors:")
print(num_vector - c(1, 1, 1, 1, 1))

# 3. Mnożenie wektorów
print("Multiplying vectors:")
print(num_vector * c(2, 2, 2, 2, 2))

# 4. Dzielenie wektorów
print("Dividing vectors:")
print(num_vector / c(2, 2, 2, 2, 2))
```

#### 8) I love Dataframes

Podstawy:
W języku R mamy różne sposoby przechowywania i manipulowania danymi w formie tabeli. 

Najpopularniejszymi i najczęściej używanymi są data.frame, data.table i tibble.

1. data.frame:
- Co to jest? To jest podstawowy, wbudowany w R sposób na przechowywanie danych w formie tabeli.
- Kiedy używać? Zawsze, gdy pracujesz z R i potrzebujesz tabeli.
- Jakie pakiety? Jest wbudowany, więc nie potrzebujesz dodatkowego pakietu. Ale dplyr (część tidyverse) pozwala na bardziej zaawansowane operacje na data.frame.

2. tibble:
- Co to jest? To nowocześniejsza wersja data.frame. Ma lepszą wydrukowaną wersję w konsoli i pewne dodatkowe funkcjonalności.
- Kiedy używać? Gdy pracujesz z tidyverse, ponieważ wiele funkcji z tidyverse zwraca tibble zamiast tradycyjnych data.frame.
- Jakie pakiety? Potrzebujesz pakietu tibble, ale jeśli zainstalujesz i załadujesz tidyverse, dostaniesz tibble automatycznie.

3. data.table:
- Co to jest? To bardziej zaawansowana wersja data.frame, optymalizowana dla szybkości i dużych danych.
- Kiedy używać? Gdy masz duże zestawy danych lub potrzebujesz bardzo szybkich operacji na danych.
- Jakie pakiety? Potrzebujesz pakietu data.table.


```r
install.packages(c("tibble", "data.table"))

library(stats)
library(tibble)
library(data.table)
```

```r
# różnice syntaksowe

# data.frame:
library(stats) # dla data.frame

# Tworzenie:
df <- data.frame(wiek = c(20, 25, 30), imię = c("Anna", "Piotr", "Zosia"))

# Odwoływanie się do kolumn:
df$wiek

# Dodawanie kolumny:
df$miasto <- c("Warszawa", "Poznań", "Gdańsk")


# tibble:
library(tibble) # dla tibble

# Tworzenie:
tb <- tibble(wiek = c(20, 25, 30), imię = c("Anna", "Piotr", "Zosia"))

# Odwoływanie się do kolumn:
tb$wiek

# Dodawanie kolumny:
tb$miasto <- c("Warszawa", "Poznań", "Gdańsk")


# data.table:
library(data.table) # dla data.table

# Tworzenie:
dt <- data.table(wiek = c(20, 25, 30), imię = c("Anna", "Piotr", "Zosia"))

# Odwoływanie się do kolumn:
dt[, wiek]

# Dodawanie kolumny:
dt[, miasto := c("Warszawa", "Poznań", "Gdańsk")]
```

#### 10) funkcje na df w R vs Python

| Operacja | R | Python (pandas) |

| Operacja | R | Python (pandas) |
|:--------:|:--------------:|:---------------:|

**Różnice:**

- **Składnia**: W R funkcje agregujące są zazwyczaj funkcjami globalnymi i przyjmują kolumnę jako argument. W Pythonie, są to metody obiektów DataFrame lub Series.
- **Braki danych (NA)**: W R wiele funkcji agregujących domyślnie nie uwzględnia wartości NA, ale można je uwzględnić, stosując dodatkowy argument `na.rm=TRUE`. W pandas, wartości NA są domyślnie pomijane podczas agregacji.
- **Typy danych**: R i Python mają różne podejścia do typów danych. Na przykład, w R, kategorie są reprezentowane jako "faktory", podczas gdy w Pythonie można używać typu "category" w pandas.

#### 12) Obsługa Nan i duplikatów w R vs Pandas

##### Brakujące wartości (NaN/NA)

| Operacja | R | Python (pandas) |
|:--------:|:---------------:|:---------------:|
| Wykrywanie NA | `is.na(df$col)` | `df['col'].isna()` |
| Zamiana NA na wartość | `df$col[is.na(df$col)] <- value` | `df['col'].fillna(value)` |
| Usunięcie wierszy z NA | `na.omit(df)` | `df.dropna()` |
| Liczba NA w kolumnie | `sum(is.na(df$col))` | `df['col'].isna().sum()` |

##### Duplikaty

| Operacja | R | Python (pandas) |
|:--------:|:---------------:|:---------------:|
| Wykrywanie duplikatów | `duplicated(df$col)` | `df['col'].duplicated()` |
| Usunięcie duplikatów | `df[!duplicated(df$col), ]` | `df.drop_duplicates(subset='col')` |
| Liczba unikalnych wartości | `length(unique(df$col))` | `df['col'].nunique()` |

**Różnice:**

- **Składnia**: Podobnie jak w przypadku funkcji agregujących, składnia jest inna między R a pandas. W R używa się globalnych funkcji, które przyjmują kolumny jako argumenty, podczas gdy w pandas metody są stosowane do obiektów DataFrame lub Series.
- **Braki danych**: W R brakujące wartości są reprezentowane jako `NA`, podczas gdy w pandas używa się `NaN` (z biblioteki numpy). W R, aby obsłużyć wartości `NA`, często trzeba użyć funkcji `na.omit` lub podobnych. W pandas, metoda `dropna()` jest często używana do usuwania wierszy z `NaN`.
- **Duplikaty**: Podczas gdy zarówno R, jak i pandas mają funkcje do wykrywania i usuwania duplikatów, składnia różni się między nimi. W R funkcja `duplicated()` zwraca wartości logiczne wskazujące duplikaty, podczas gdy w pandas metoda `duplicated()` robi to samo.


```r
## Brakujące wartości (NA) i Duplikaty w R

### Brakujące wartości (NA)


# Załadowanie przykładowych danych z NA
data <- data.frame(a = c(1, 2, NA, 4, 5), b = c(NA, 2, 3, 4, NA))

# Wykrywanie NA
is.na(data$a)

# Zamiana NA na wartość (np. 0)
data$a[is.na(data$a)] <- 0

# Usunięcie wierszy z NA
na.omit(data)

# Liczba NA w kolumnie
sum(is.na(data$a))
```

```r
# Przykładowe dane z duplikatami
data_dup <- data.frame(a = c(1, 2, 2, 4, 5))

# Wykrywanie duplikatów
duplicated(data_dup$a)

# Usunięcie duplikatów
data_clean <- data_dup[!duplicated(data_dup$a), ]

# Liczba unikalnych wartości
length(unique(data_dup$a))
```

```r
library(dplyr)

# Zamiana NA na wartość (np. 0) z użyciem `mutate()`
data %>% mutate(a = ifelse(is.na(a), 0, a))

# Liczba NA w kolumnie z użyciem `summarise()`
data %>% summarise(na_count = sum(is.na(a)))

# Usunięcie duplikatów z użyciem `distinct()`
data_dup %>% distinct(a, .keep_all = TRUE)
```
